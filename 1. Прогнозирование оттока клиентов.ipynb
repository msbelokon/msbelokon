# Шаг 3. Постройте модель прогнозирования оттока клиентов<a name="3"></a>

Исключим признаки с мультиколлениарностью

```python
df = df.drop(columns=['Contract_period', 'Avg_class_frequency_total'])
```

Разбиение данных на обучающую и валидационную выборки функцией

```python
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=0)
```

Стандартизируем данные

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

Обучение модели логистической регрессией

```python
model = LogisticRegression(solver='liblinear')
```

```python
model.fit(X_train_st, y_train);
```

Решение модели логистической регрессии

```python
lr_predictions = model.predict(X_test_st)
lr_probabilities = model.predict_proba(X_test_st)[:,1]
```

Обучение модели случайный лес

```python
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)
rf_model.fit(X_train_st, y_train);
```

Решение модели случайный лес

```python
rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:, 1]
```

Напишем функцию для вывода метрик

```python
def print_all_metrics(y_true, y_pred, y_proba, title='Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred)))
    print('\tROC_AUC: {:.2f}'.format(roc_auc_score(y_true, y_proba)))
```

```python
def all_metrics(y_true, y_pred, y_proba):
    acc = accuracy_score(y_true, y_pred)
    pre = precision_score(y_true, y_pred)
    re = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    RC = roc_auc_score(y_true, y_proba)
    return acc, pre, re, f1, RC
```

Выведем метрики

```python
print_all_metrics(
    y_test,
    lr_predictions,
    lr_probabilities,
    title='Метрики для модели логистической регрессии:',
);
```

```python
print_all_metrics(
    y_test,
    rf_predictions,
    rf_probabilities,
    title='Метрики для модели случайного леса:'
)
```

Создадим датафрейм с метриками для наглядного сравнения

```python
names = ['Accuracy', 'Precision', 'Recall', 'F1', 'ROC_AUC']
```

```python
lr = list(all_metrics(
    y_test,
    lr_predictions,
    lr_probabilities))
```

```python
rf = list(all_metrics(
    y_test,
    rf_predictions,
    rf_probabilities))
```

```python
res = pd.DataFrame({'metric': names, 'LogReg': lr, 'RanFor': rf})
```

```python
res['best_model'] = ['LogReg' if x==True else 'RanFor' for x in res['LogReg'] > res['RanFor']]
```

```python
res
```

Чем ближе к 1, тем лучше показатель модели.  
Показатели обеих моделей хорошие, т.к близки к 1.  
    
Доля правильных ответов и точность лучше у модели случайный лес, а вот по критериям полнота, F-мера и площадь под кривой лучшие результаты показала модель логистической регрессии.
    
т.к. сводные показатели лучше у модели логистической регрессии, то предпочтение отдадим модели логистической регрессии.

# Шаг 4. Сделайте кластеризацию клиентов<a name="4"></a>

Стандартизируем данные

```python
sc = StandardScaler()
x_sc = sc.fit_transform(df.drop(columns=['Churn']))
```

Построим матрицу расстояние и выведем дендрограмму

```python
linked = linkage(x_sc, method = 'ward') 
```

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Hierarchial clustering')
plt.show() 
```

<div style="border:solid purple 2px; padding: 20px">    

Дендрограмма предлагает 4 итоговых кластера, но т.к. по условию работа должна выполняться на 5 кластерах, проведем кластеризацию с количеством кластеров = 5
    
</div>

Обучим модель кластеризации на основании алгоритма K-Means с количеством кластеров = 5

```python
km = KMeans(n_clusters=5, random_state=0)
```

```python
labels = km.fit_predict(x_sc)
```

```python
df['cluster'] = labels
```

Выведем группировку средних значений по кластерам
